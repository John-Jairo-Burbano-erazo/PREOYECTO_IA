```python
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score

data=pd.read_excel('/content/si-sustentabilidad-john - copia.xlsx')

#asignamos nuestra variable dy.e entreda por entranamieneto y las etiquetas y.
datax=[["superficie_atendida_con_manejo_integral_de_plagas"]]
x_train= data[["superficie_atendida_con_manejo_integral_de_plagas"]]
y_train=data["superficie_atendida_con_control_biologico"].values

# Create a linear regression model
model = linear_model.LinearRegression()

# Fit the model to the training data
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_train)

# Print the coefficients, intercept, mean squared error, and variance score
print('Coefficients: \n', model.coef_)
print('Intercept: \n', model.intercept_)
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))
print('Variance score: %.2f' % r2_score(y_train, y_pred))
```

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(x_train, y_train, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(x_train, y_pred, color='red', linewidth=2, label='Regression Line')

# Add labels and title
plt.xlabel('Superficie atendida con manejo integral de plagas')
plt.ylabel('Superficie atendida con control biologico')
plt.title('Linear Regression Analysis')
plt.legend()
plt.show()
```
